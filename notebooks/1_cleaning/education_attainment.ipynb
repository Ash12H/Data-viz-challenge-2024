```python
import pandas as pd
import plotly.express as px
from dash import dash_table

pd.options.plotting.backend = "plotly"
```

```python
data = pd.read_csv("../../data/1_raw/education_attainment.csv", sep=";")
```

## Remove empty


```python
data = data.dropna(axis="columns", how="all")
data.info()
```

```python
columns = [
    "Pays et territoires insulaires du Pacifique",
    "Sexe",
    "Âge",
    "Urbanisation",
    "Invalidité",
    "Niveau d'éducation",
    "TIME_PERIOD",
    "OBS_VALUE",
]
data = data[columns]
```

```python
data = data.replace(
    {
        "Îles Marshall (pays)": "Marshall Islands",
        "Palaos": "Palau",
    }
)
```

```python
list(data["Pays et territoires insulaires du Pacifique"].unique())
```

```python
for i in [
    "Pays et territoires insulaires du Pacifique",
    "Sexe",
    "Âge",
    "Urbanisation",
    "Invalidité",
    "Niveau d'éducation",
]:
    data[i] = data[i].astype("category")
```

```python
data.describe(include="all")
```

```python
data.info()
```

```python
data = (
    data.pivot_table(
        index=[
            "Pays et territoires insulaires du Pacifique",
            "Sexe",
            "Âge",
            "Urbanisation",
            "Invalidité",
            "TIME_PERIOD",
        ],
        columns="Niveau d'éducation",
        values="OBS_VALUE",
    )
    .sort_index()
    .reset_index()
)
data = data[data.columns[data.notna().sum() > 10]]
data.head()
```

```python
data.to_parquet("../../data/2_clean/education_attainment.parquet")
```
