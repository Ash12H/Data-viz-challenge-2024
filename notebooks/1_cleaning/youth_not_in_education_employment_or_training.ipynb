```python
import pandas as pd
import plotly.express as px
from dash import dash_table

pd.options.plotting.backend = "plotly"
```

```python
data = pd.read_csv(
    "../../data/1_raw/youth_not_in_education_employment_or_training.csv", sep=";"
)
```

## Remove empty


```python
data = data.dropna(axis="columns", how="all")
data.info()
```

```python
columns = [
    "Pays et territoires insulaires du Pacifique",
    "Indicateur",
    "Sexe",
    "Urbanisation",
    "Invalidité",
    "TIME_PERIOD",
    "OBS_VALUE",
]
data = data[columns]
```

```python
list(data["Pays et territoires insulaires du Pacifique"].unique())
```

```python
data = data.replace(
    {
        "Îles Marshall (pays)": "Marshall Islands",
        "Salomon": "Solomon Islands",
        "Palaos": "Palau",
        "Samoa": "Papua New Guinea",
        "Îles Cook": "Cook Islands",
        "États fédérés de Micronésie (pays)": "Micronesia",
    }
)
```

```python
for i in [
    "Pays et territoires insulaires du Pacifique",
    "Indicateur",
    "Sexe",
    "Urbanisation",
    "Invalidité",
]:
    data[i] = data[i].astype("category")
```

```python
data.describe(include="all")
```

```python
data.info()
```

```python
data = data.pivot_table(
    index=[
        "Pays et territoires insulaires du Pacifique",
        "Sexe",
        "Urbanisation",
        "Invalidité",
        "TIME_PERIOD",
    ],
    columns="Indicateur",
    values="OBS_VALUE",
).reset_index()
data = data[data.columns[data.notna().sum() > 10]]
data.head()
```

```python
data.to_parquet(
    "../../data/2_clean/youth_not_in_education_employment_or_training.parquet"
)
```
