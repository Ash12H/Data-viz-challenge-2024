```python
import pandas as pd

pd.options.plotting.backend = "plotly"
```

```python
data = pd.read_csv("../../data/1_raw/literacy_rates.csv", sep=";")
```

## Remove empty


```python
data = data.dropna(axis="columns", how="all")
data.info()
```

```python
columns = [
    "Pays et territoires insulaires du Pacifique",
    "Indicateur",
    "Sexe",
    "Âge",
    "Urbanisation",
    "Invalidité",
    "TIME_PERIOD",
    "OBS_VALUE",
]
data = data[columns]
```

```python
data.head()
```

```python
for i in [
    "Pays et territoires insulaires du Pacifique",
    "Sexe",
    "Âge",
    "Urbanisation",
    "Invalidité",
    "Indicateur",
]:
    data[i] = data[i].astype("category")
```

```python
data.describe(include="all")
```

```python
# Je souhaite transformer les valeurs unique de la colonne "Indicateur" en colonnes
# et les valeurs de la colonne "OBS_VALUE" en valeurs
data = (
    data.pivot_table(
        index=[
            "Pays et territoires insulaires du Pacifique",
            "Sexe",
            "Âge",
            "Urbanisation",
            "Invalidité",
            "TIME_PERIOD",
        ],
        columns="Indicateur",
        values="OBS_VALUE",
    )
    .dropna(axis="rows", how="all")
    .reset_index()
)
data = data[data.columns[data.notna().sum() > 10]]
data.head()
```

```python
data.to_parquet("../../data/2_clean/literacy_rates.parquet")
```
