```python
import geopandas as gpd
import shapely
```

## Kiribati


```python
kiribati = gpd.read_file("../../data/shapes/1_raw/kiribati.geojson")
kiribati = kiribati.translate(
    xoff=-float(kiribati.centroid.x), yoff=-float(kiribati.centroid.y)
)
kiribati.plot()
```

## Samoa


```python
samoa = gpd.read_file("../../data/shapes/1_raw/samoa.geojson")

# merge 2 geometrys in a multi polygon
samoa = samoa.dissolve()
samoa = samoa.translate(xoff=-float(samoa.centroid.x), yoff=-float(samoa.centroid.y))
samoa.plot()
```
