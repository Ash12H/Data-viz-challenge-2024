```python
import shapely
import geopandas as gpd
import pandas as pd
```

```python
eez = gpd.read_file("../../data/shapes/1_raw/country_boundary_eez.geojson")
```

```python
eez
```

```python
sovereigns = {
    "Tonga": "Tonga",
    "Tuvalu": "Tuvalu",
    "Marshall Islands": "Îles Marshall",
    "Samoa": "Samoa",
    "Micronesia": "États fédérés de Micronésie",
    "Vanuatu": "Vanuatu",
    "Nauru": "Nauru",
    "Kiribati": "Kiribati",
    "Solomon Islands": "Salomon",
    "Palau": "Palaos",
    "Papua New Guinea": "Papouasie-Nouvelle-Guinée",
    "Fiji": "Fidji",
}
territories = {
    "Cook Islands": "Îles Cook",
    "Niue": "Niue",
    "Tokelau": "Tokelau",
    "Wallis and Futuna": "Wallis-et-Futuna",
    "New Caledonia": "Nouvelle-Calédonie",
    "French Polynesia": "Polynésie française",
    "Northern Mariana Islands": "Îles Mariannes du Nord",
}
```

```python
eez = eez[
    eez.isin({"Sovereign1": sovereigns.keys(), "Territory1": territories.keys()})
    .sum(axis=1)
    .astype(bool)
]
```

```python
print(eez["Sovereign1"].unique())
print(eez["Territory1"].unique())
```

```python
kiribati_geo = (
    eez.query("Sovereign1 == 'Kiribati'")
    .groupby("Sovereign1")["geometry"]
    .agg(lambda x: x.unary_union)
)
caledonia_geo = (
    eez.query("Territory1 == 'New Caledonia'")
    .groupby("Territory1")["geometry"]
    .agg(lambda x: x.unary_union)
)
eez = eez.drop(eez.query("Sovereign1 == 'Kiribati'").index)
eez = eez.drop(eez.query("Territory1 == 'New Caledonia'").index)
eez
```

```python
sov_geo = eez.groupby("Territory1")["geometry"].agg(lambda x: x.unary_union)
sov_geo = pd.concat((sov_geo, kiribati_geo, caledonia_geo))
sov_geo.index = sov_geo.index.rename("pacific_island")
pacific_eez = sov_geo.reset_index()
pacific_eez
```
